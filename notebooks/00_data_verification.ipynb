## Dataset check

````python
DATA_URL = (
    'https://data.open-power-system-data.org/time_series/'
    '2020-10-06/time_series_60min_singleindex.csv'
)

needed = [
    'utc_timestamp',
    'DE_load_actual_entsoe_transparency',
    'DE_load_forecast_entsoe_transparency',
]

# Load only a small sample to keep this check quick.
sample = pd.read_csv(DATA_URL, usecols=needed, nrows=500)
sample.head()
````

````python
# Verify required columns are present.
for col in needed:
    present = col in sample.columns
    print(f'{"OK " if present else "MISSING"}  {col}')
````

````python
# Data types
sample.dtypes
````

````python
# Missing values in the small sample
sample.isna().sum()
````

````python
# Date range of the sample
ts = pd.to_datetime(sample['utc_timestamp'], utc=True)
print('First timestamp:', ts.min())
print('Last timestamp :', ts.max())
````

## Quick plot

````python
df = sample.copy()
df['timestamp'] = pd.to_datetime(df['utc_timestamp'], utc=True)
df = df.rename(columns={'DE_load_actual_entsoe_transparency': 'actual_load_MW'})

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df['timestamp'], df['actual_load_MW'], color='steelblue')
ax.set_xlabel('Time')
ax.set_ylabel('Actual load (MW)')
ax.set_title('Quick look at the first hours of the dataset')
fig.autofmt_xdate()
plt.show()
````
